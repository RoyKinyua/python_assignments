```python
###List of prime numbers using list comprehension
 
primes = [n for n in range(2, 51) if all(n % i != 0 for i in range(2, int(n*0.5)+1))]
 
print("Prime numbers from 2 to 50:", primes)
```

```python
###Prime numbers between 1 and 50
 
for num in range(2, 51): 
    is_prime = True
    for i in range(2, int(num*0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
```

```python
#class employee with name and salary attributes then create a subclass with an extra department attribute and method to display full details
class Employee :
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def speak(self):
        return f"The name of the employee is {self.name} and earns {self.salary}"
```

```python
class Manager(Employee):
    def __init__(self,name,salary,department):
     super().__init__(name,salary) #call the parent constructor
     self.department=department
    def speak(self):
            return f"The manager is called {self.name}, earns a salary of {self.salary} and is in the {self.department} department."
```

```python
manager1=Manager("Roy",20000,"IT")
print(manager1.speak())
```

#### Decorators
In Python, decorators are flexible way to modify or extend behavior of functions or methods, without changing their actual code.

A decorator is essentially a function that takes another function as an argument and returns a new function with enhanced functionality.
Decorators are often used in scenarios such as logging, authentication and memorization, allowing us to add additional functionality to existing functions or methods in a clean, reusable way.

```python
#list of fruits
fruits={"mango","apple","orange","pineapple","watermelon"}
print(fruits)
```

```python
for index, fruit in enumerate(fruits, start=1):
    print(f"Fruit {index}: {fruit}")
    if index == 3:
        print("Reached the third fruit. Stopping the loop.")
        break
```
